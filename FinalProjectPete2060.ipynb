```python
Rst = df['Rst']
SGgas = df['SGgas']
plt.scatter(Rst, SGgas, edgecolors ='r')
plt.xlabel('SGgas')
plt.ylabel('Rst')
plt.title('Rst vs SGgas')
plt.show()
```

```python
Rst = df['Rst']
Psp = df['Psp']
plt.scatter(Rst, Psp, edgecolors ='m')
plt.xlabel('Psp')
plt.ylabel('Rst')
plt.title('Rst vs Psp')
plt.show()
```

```python
df['log_Psp']=np.log(df.Psp)
df['log_Tsp']=np.log(df.Tsp)
df['log_Rst']=np.log(df.Rst)
df['log_SGoil']=np.log(df.SGoil)
df['log_SGgas']=np.log(df.SGgas)
df.head()
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
from sklearn import linear_model
```

```python
X = df[['log_SGoil','log_SGgas','log_Psp','log_Tsp']]
Y = df['log_Rst']
regr = linear_model.LinearRegression()
regr.fit(X, Y)
```

```python
print(regr.intercept_)
print(regr.coef_) 
model = sm.OLS(Y, X).fit()
predictions = model.predict(X) 
print_model = model.summary()
print(print_model)
```
